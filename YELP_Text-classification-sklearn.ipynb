```
from textblob import TextBlob
```

```
def get_sentiment(text):
  pred = TextBlob(text).sentiment[0]
  if pred > 0:
    return 1
  else:
    return 0
```

```
test.text.apply(get_sentiment)
```

```
print(metrics.classification_report(test.sentiment, test.text.apply(get_sentiment)))
```
